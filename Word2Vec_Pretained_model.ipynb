```python
import gensim
from gensim.models import Word2Vec, KeyedVectors
```

```python
# !pip install wget
# !wget -c "https://s3.amazonaws.com/dl4j-distribution/GoogleNews-vectors-negative300.bin.gz"
```

```python
# Pre-trained model from google
model = KeyedVectors.load_word2vec_format('./google/GoogleNews-vectors-negative300.bin.gz', binary=True)
```

```python
model.most_similar('mango')
```

```python
model.most_similar('Jatrabari')
```

```python
# similarity
model.similarity('man', 'women')
```

```python
model.similarity('man', 'water')
```

```python
model.doesnt_match(['man','boy','kid'])
```

```python
model.doesnt_match(['python','mango','c++'])
```

```python
# Arithmetic operation
res = model['king'] - model['man'] + model['women']
model.most_similar([res])
```

```python
res = model['TK'] - model['Bangladesh'] + model['USA']
model.most_similar([res])
```
