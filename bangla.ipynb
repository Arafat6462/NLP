```python
story = []
for sent in corpus:
#     story.append(simple_preprocess(sent, min_len=2))
    story.append(word_tokenize(sent))

    
print(type(story))
print(len(story))
print(story[:10])
```

```python
story[:5]
```

```python
# model building
model = gensim.models.Word2Vec(window=4)
model.build_vocab(story)
model.train(story, total_examples=model.corpus_count, epochs=model.epochs)
```
