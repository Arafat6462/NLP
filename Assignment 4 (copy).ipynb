# Lowercasing

```python
data['clean_review'] = data['review'].str.lower()
# re-ordering column
data = data[['review', 'clean_review', 'sentiment']]
data.head(50)
```

# Removing HTML tags


```python
def remove_html_tags(reviews):
    # re.sub(pattern, repl, string, count=0, flags=0)
    result = re.sub('<.*?>', '', reviews)
    return result
```

```python
data['clean_review'] = data['clean_review'].apply(
    lambda x: remove_html_tags(x))
data.head(50)
```

# 4. Remove Punctuation

```python
string.punctuation
```

```python
def remove_punctuation(review):
#     review = review.translate(str.maketrans("", "", string.punctuation))
    review = re.sub(pattern="[.]", repl=' ', string=review)
    review = re.sub(pattern="[^\w\s]", repl='', string=review)    
    return review
```

```python
# text = "hello   ara'fa,t. how.are you. wh.a.t are y'o'u doning"
# remove_punctuation(text)
```

```python
data['clean_review'] = data['clean_review'].apply(
    lambda x: remove_punctuation(x))
data.head(50)
```

# Removing Stop words


```python
from nltk.corpus import stopwords
', '.join(stopwords.words('english'))
```

```python
STOPWORDS = set(stopwords.words('english'))
STOPWORDS
```

```python
def remove_stopwords(review):
    return " ".join([word for word in review.split() if word not in STOPWORDS])
```

```python
data['clean_review'] = data['clean_review'].apply(lambda x: remove_stopwords(x))
data.sample(3)
```

#  Tokenization

```python
from nltk import sent_tokenize, word_tokenize

def tokenization(review):
    word = word_tokenize(review)
    return word 
```

```python
data['clean_review'] = data['clean_review'].apply(lambda x: tokenization(x))
data.head()
```

# Lemmatization

```python
from nltk.stem import WordNetLemmatizer
lemma = WordNetLemmatizer()
```

```python
def lemmatization(review):
    return " ".join([lemma.lemmatize(word) for word in review])
```

```python
data['lemmatize'] = data['clean_review'].apply(lambda x: lemmatization(x))
data.head()
```

#### Remove extra columns

```python
data = data[['lemmatize', 'sentiment']]
data.head()
```

# sentiment column str to int

```python
def sentiment(senti):
    if senti == 'positive':
        return 1
    elif senti == 'negative':
        return 0
```

```python
data['sentiment'] = data['sentiment'].apply(lambda x: sentiment(x))
```

# Find out the number of words in the entire corpus and also the total number of unique words(vocabulary) using just python

```python
# total number in corpus
def total_words(review):
    count = 0
    for sententse in review:
        count += len(sententse.split())
    return count
```

```python
total_word = total_words(data['lemmatize'])
print(f"Total word is corpus : {total_word}")
```
